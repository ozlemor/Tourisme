```python
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

sns.set_theme(style='whitegrid')
plt.rcParams['figure.figsize'] = (12, 5)

print('Bibliothèques importées avec succès')
```

```python
# Chargement des 6 fichiers sources
df_dest     = pd.read_csv('data/01_destinations_brut.csv')
df_signaux  = pd.read_excel('data/02_signaux_marche.xlsx')
df_facteurs = pd.read_csv('data/04_facteurs_externes.csv', sep=';')
df_cible    = pd.read_csv('data/05_cible_recommandation.csv')

with open('data/03_reviews_reseaux.json', encoding='utf-8') as f:
    df_reviews = pd.DataFrame(json.load(f))

with open('data/06_campaign_history.json', encoding='utf-8') as f:
    df_campaign = pd.DataFrame(json.load(f))

print('=== CHARGEMENT TERMINÉ ===')
print(f'01_destinations  : {df_dest.shape}')
print(f'02_signaux       : {df_signaux.shape}')
print(f'04_facteurs      : {df_facteurs.shape}')
print(f'03_reviews       : {df_reviews.shape}')
print(f'06_campaign      : {df_campaign.shape}')
print()
print('05_cible contenu :')
print(df_cible)
```

```python
# Dimensions et types — 01_destinations
print('=== 01_destinations_brut ===')
print(f'Dimensions : {df_dest.shape[0]} lignes x {df_dest.shape[1]} colonnes')
print()
for col in df_dest.columns:
    print(f'  {col:20} → {str(df_dest[col].dtype):10} | exemple : {df_dest[col].iloc[0]}')
print()
df_dest.head()
```

```python
# Dimensions et types — 02_signaux
print('=== 02_signaux_marche ===')
print(f'Dimensions : {df_signaux.shape[0]} lignes x {df_signaux.shape[1]} colonnes')
print(f'Pays       : {sorted(df_signaux["country"].unique())}')
print(f'Période    : {df_signaux["month"].min()} → {df_signaux["month"].max()}')
print()
df_signaux.head()
```

```python
# Dimensions et types — 03_reviews
print('=== 03_reviews_reseaux ===')
print(f'Dimensions : {df_reviews.shape}')
print(f'Sentiments : {df_reviews["sentiment"].unique()}')
print(f'Score min/max : {df_reviews["score"].min()} / {df_reviews["score"].max()}')
print()
df_reviews.head()
```

```python
# Dimensions et types — 04_facteurs
print('=== 04_facteurs_externes ===')
print(f'Dimensions : {df_facteurs.shape}')
print(f'flight_price : {df_facteurs["flight_price"].unique()}')
print(f'weather      : {df_facteurs["weather"].unique()}')
print()
df_facteurs.head()
```

```python
# Dimensions et types — 06_campaign
print('=== 06_campaign_history ===')
print(f'Dimensions : {df_campaign.shape}')
print(f'Status     : {df_campaign["status"].unique()}')
print()
df_campaign
```
